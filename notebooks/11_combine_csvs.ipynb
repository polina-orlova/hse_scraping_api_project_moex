```python
import pandas as pd
```

```python
df_SBER = pd.read_csv('moex_csvs/candles_SBER.csv')
df_GAZP = pd.read_csv('moex_csvs/candles_GAZP.csv')
df_LKOH = pd.read_csv('moex_csvs/candles_LKOH.csv')
df_ROSN = pd.read_csv('moex_csvs/candles_ROSN.csv')
df_MTSS = pd.read_csv('moex_csvs/candles_MTSS.csv')
df_GMKN = pd.read_csv('moex_csvs/candles_GMKN.csv')
df_MGNT = pd.read_csv('moex_csvs/candles_MGNT.csv')
df_ALRS = pd.read_csv('moex_csvs/candles_ALRS.csv')
df_VTBR = pd.read_csv('moex_csvs/candles_VTBR.csv')

df_vse_candles = pd.concat([df_SBER, df_GAZP, df_LKOH, df_ROSN, df_MTSS, df_GMKN, df_MGNT, df_ALRS, df_VTBR], ignore_index=True)
df_vse_candles['begin'] = pd.to_datetime(df_vse_candles['begin'])
df_vse_candles
```

```python
df_infa_SBER = pd.read_csv('moex_csvs/infa_SBER.csv')
df_infa_GAZP = pd.read_csv('moex_csvs/infa_GAZP.csv')
df_infa_LKOH = pd.read_csv('moex_csvs/infa_LKOH.csv')
df_infa_ROSN = pd.read_csv('moex_csvs/infa_ROSN.csv')
df_infa_MTSS = pd.read_csv('moex_csvs/infa_MTSS.csv')
df_infa_GMKN = pd.read_csv('moex_csvs/infa_GMKN.csv')
df_infa_MGNT = pd.read_csv('moex_csvs/infa_MGNT.csv')
df_infa_ALRS = pd.read_csv('moex_csvs/infa_ALRS.csv')
df_infa_VTBR = pd.read_csv('moex_csvs/infa_VTBR.csv')

df_vse_infa = pd.concat([df_infa_SBER, df_infa_GAZP, df_infa_LKOH, df_infa_ROSN, df_infa_MTSS, df_infa_GMKN, df_infa_MGNT, df_infa_ALRS, df_infa_VTBR], ignore_index=True)
df_vse_infa
```

```python
df_diva_SBER = pd.read_csv('moex_csvs/diva_SBER.csv')
df_diva_GAZP = pd.read_csv('moex_csvs/diva_GAZP.csv')
df_diva_LKOH = pd.read_csv('moex_csvs/diva_LKOH.csv')
df_diva_ROSN = pd.read_csv('moex_csvs/diva_ROSN.csv')
df_diva_MTSS = pd.read_csv('moex_csvs/diva_MTSS.csv')
df_diva_GMKN = pd.read_csv('moex_csvs/diva_GMKN.csv')
df_diva_MGNT = pd.read_csv('moex_csvs/diva_MGNT.csv')
df_diva_ALRS = pd.read_csv('moex_csvs/diva_ALRS.csv')
df_diva_VTBR = pd.read_csv('moex_csvs/diva_VTBR.csv')

df_vse_diva = pd.concat([df_diva_SBER, df_diva_GAZP, df_diva_LKOH, df_diva_ROSN, df_diva_MTSS, df_diva_GMKN, df_diva_MGNT, df_diva_ALRS, df_diva_VTBR], ignore_index=True)
df_vse_diva['registryclosedate'] = pd.to_datetime(df_vse_diva['registryclosedate'])
df_vse_diva
```

```python
df_markdata_SBER = pd.read_csv('moex_csvs/markdata_SBER.csv')
df_markdata_GAZP = pd.read_csv('moex_csvs/markdata_GAZP.csv')
df_markdata_LKOH = pd.read_csv('moex_csvs/markdata_LKOH.csv')
df_markdata_ROSN = pd.read_csv('moex_csvs/markdata_ROSN.csv')
df_markdata_MTSS = pd.read_csv('moex_csvs/markdata_MTSS.csv')
df_markdata_GMKN = pd.read_csv('moex_csvs/markdata_GMKN.csv')
df_markdata_MGNT = pd.read_csv('moex_csvs/markdata_MGNT.csv')
df_markdata_ALRS = pd.read_csv('moex_csvs/markdata_ALRS.csv')
df_markdata_VTBR = pd.read_csv('moex_csvs/markdata_VTBR.csv')

df_vse_markdata = pd.concat([df_markdata_SBER, df_markdata_GAZP, df_markdata_LKOH, df_markdata_ROSN, df_markdata_MTSS, df_markdata_GMKN, df_markdata_MGNT, df_markdata_ALRS, df_markdata_VTBR], ignore_index=True)
df_vse_markdata
```

```python
df_idishka_SBER = pd.read_csv('moex_csvs/idishka_SBER.csv')
df_idishka_GAZP = pd.read_csv('moex_csvs/idishka_GAZP.csv')
df_idishka_LKOH = pd.read_csv('moex_csvs/idishka_LKOH.csv')
df_idishka_ROSN = pd.read_csv('moex_csvs/idishka_ROSN.csv')
df_idishka_MTSS = pd.read_csv('moex_csvs/idishka_MTSS.csv')
df_idishka_GMKN = pd.read_csv('moex_csvs/idishka_GMKN.csv')
df_idishka_MGNT = pd.read_csv('moex_csvs/idishka_MGNT.csv')
df_idishka_ALRS = pd.read_csv('moex_csvs/idishka_ALRS.csv')
df_idishka_VTBR = pd.read_csv('moex_csvs/idishka_VTBR.csv')

df_vse_idishka = pd.concat([df_idishka_SBER, df_idishka_GAZP, df_idishka_LKOH, df_idishka_ROSN, df_idishka_MTSS, df_idishka_GMKN, df_idishka_MGNT, df_idishka_ALRS, df_idishka_VTBR], ignore_index=True)
df_vse_idishka
```

```python
df_mega = df_vse_candles.merge(df_vse_infa, on='ticker')
df_mega = df_mega.merge(df_vse_markdata, on='ticker')
df_mega
```

```python
df_vse_diva.to_csv('vse_dividends.csv', index=False)
df_vse_idishka.to_csv('vse_indices.csv', index=False)
df_mega.to_csv('mega_dataset.csv', index=False) #тут у нас candels, infa, marketdata
```
